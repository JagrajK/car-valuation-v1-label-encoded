```python
## import libraries
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from torch import nn
```

```python
##set up data
df = pd.read_csv(r"C:\Users\Jagraj\Coding\pytorch, im that guy pal\datasets\clean_data.csv", encoding="ISO-8859-1")
```

```python
df = df.sort_values(by='model')
```

```python
df['encoded_model'] = df['model'].astype('category').cat.codes
model_key = df[['model', 'encoded_model']].drop_duplicates().set_index('encoded_model')

df['encoded_transmission'] = df['transmission_type'].astype('category').cat.codes
transmission_key = df[['transmission_type', 'encoded_transmission']].drop_duplicates().set_index('encoded_transmission')

df['encoded_fuel'] = df['fuel_type'].astype('category').cat.codes
fuel_key = df[['fuel_type', 'encoded_fuel']].drop_duplicates().set_index('encoded_fuel')
```

```python
df = df.drop(columns = ['Unnamed: 0', 'brand', 'registration_date', 'power_ps', 'fuel_consumption_g_km', 'offer_description', 'model', 'color', 'transmission_type', 'fuel_type', 'fuel_consumption_l_100km'], axis = 1)
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
df.drop("price_in_euro", axis = 1)
```

```python
df.dtypes
```

```python
Y = torch.tensor(df.price_in_euro)

columns_to_concat = df.drop(columns = "price_in_euro", axis = 1).columns
column_values = []
'''for column in columns_to_concat:
    a1 = df[column].values
    a2 = torch.tensor(a1)
    a3 = torch.reshape(a2, (len(df), 1))
    column_values.append(a3)'''
column_values = [torch.reshape(torch.tensor(df[column].values), (len(df),1)) for column in columns_to_concat]
X = torch.cat([values for values in column_values], dim = 1)
```

```python
print(Y.shape)
X.size()
```

```python
test_percent = 0.05
seed = 69420

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = test_percent, random_state = seed)
```

```python
b = 128

Y_train, Y_test = Y_train.type(torch.float32).to(device), Y_test.type(torch.float32).to(device)
X_train, X_test = X_train.to(device).type(torch.float32), X_test.to(device).type(torch.float32)

train_data = torch.utils.data.TensorDataset(X_train, Y_train)
test_data = torch.utils.data.TensorDataset(X_test, Y_test)

train_dataloader = torch.utils.data.DataLoader(train_data,
                                               batch_size = b,
                                               shuffle = True)

test_dataloader = torch.utils.data.DataLoader(test_data,
                                              batch_size = b,
                                              shuffle = False)
```

```python
D = X_train.size(1)
K = 1
nodes = 2 ** 10

#finally a neural net
class carlearner(nn.Module):
    def __init__(self):
        super().__init__()
        self.input = nn.Sequential(nn.Linear(D, nodes), 
                                   nn.ReLU())
        self.mid1 = nn.Sequential(nn.Linear(nodes, nodes), 
                                 nn.ReLU(),
                                nn.Linear(nodes, nodes), 
                                 nn.ReLU())
        self.mid2 = nn.Sequential(nn.Linear(nodes, nodes), 
                                 nn.ReLU(),
                                nn.Linear(nodes, nodes), 
                                 nn.ReLU())
        self.mid3 = nn.Sequential(nn.Linear(nodes, nodes), 
                                 nn.ReLU(),
                                nn.Linear(nodes, nodes), 
                                 nn.ReLU())
        self.output = nn.Linear(nodes, K)
    
    def forward(self, x):
        '''#print(x.shape)
        r1 = self.input(x)
        #print(r1.shape)
        r2 = self.mid(r1)
        if layers == 0:
            return self.output(r1).squeeze(dim=1)
        elif layers == 1:
            return self.output(r2).squeeze(dim=1)
        else:
            for i in range(layers - 1):
                r2 = self.mid(r2)
            return self.output(r2).squeeze(dim=1)
        '''
        return self.output(self.mid3(self.mid2(self.mid1(self.input(x)))))
```

```python
#instance of class
model = carlearner().to(device)
```

```python
def train_step(model: torch.nn.Module,
               data_loader: torch.utils.data.DataLoader,
               loss_fn: torch.nn.Module,
               optim: torch.optim.Optimizer,
               device: torch.device = device):
    train_loss = 0
    model.train()
    for batch, (X, y) in enumerate(data_loader):
        
        y_pred = model(X)
        # print(y.shape)
        loss = loss_fn(y_pred, y)
        #print(y_pred.shape)
        #print(loss)
        train_loss += loss.item()
        
        optim.zero_grad()
        
        loss.backward()
        
        optim.step()
        
        if batch * b % 4000 == 0:
            print(f"for the {b * batch}th data entry, cumulative loss is {train_loss/(batch+1)}, {epoch}")
#

def test_step(model: torch.nn.Module,
               data_loader: torch.utils.data.DataLoader,
               loss_fn: torch.nn.Module,
               device: torch.device = device):
    test_loss = 0
    model.eval()
    with torch.inference_mode():
        for X, y in data_loader:
            test_pred = model(X)
            test_loss += loss_fn(test_pred, y).item()
    
    test_loss /= len(data_loader)
    print(f"test loss = {test_loss}")
```

```python
#retrain with lower rate and batch size
epochs = 5
rate = 0.0005

loss = nn.L1Loss()
optim = torch.optim.Adam(params = model.parameters(), lr = rate)

for epoch in range(epochs):
    train_step(model = model,
               data_loader = train_dataloader,
               loss_fn = loss,
               optim = optim)
    
    test_step(model = model,
              data_loader = test_dataloader,
              loss_fn = loss)
```
